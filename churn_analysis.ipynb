```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Create a cross-tab of counts
ct = pd.crosstab(churn_analysis['gender'], churn_analysis['Churn'])

# 2. Convert to percentages
ct_percent = ct.div(ct.sum(axis=1), axis=0) * 100

# 3. Plot the stacked bar chart
ax = ct_percent.plot(kind='bar', stacked=True, figsize=(6, 4), colormap='Set2')

# 4. Add percentage labels
for i, row in enumerate(ct_percent.values):
    cumulative = 0
    for j, val in enumerate(row):
        if val > 0:
            ax.text(i, cumulative + val / 2, f"{val:.1f}%", ha='center', va='center', fontsize=10)
            cumulative += val

# 5. Titles and labels
plt.title("Churn Analysis by Gender")
plt.xlabel("Gender")
plt.ylabel("Percentage")
plt.legend(title='Churn')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# List of columns to plot
cols = ['PhoneService', 'MultipleLines', 'InternetService',
        'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
        'TechSupport', 'StreamingTV', 'StreamingMovies']

# Set up the grid for subplots (3 rows x 3 columns)
fig, axes = plt.subplots(3, 3, figsize=(20, 20))
fig.suptitle("Churn Analysis by Service Features", fontsize=16)

# Flatten axes array for easy iteration
axes = axes.flatten()

# Create countplots
for i, col in enumerate(cols):
    sns.countplot(x=col, data=churn_analysis, hue='Churn', ax=axes[i])
    axes[i].set_title(f"{col} vs Churn")
    axes[i].tick_params(axis='x', rotation=45)

# Adjust layout
plt.tight_layout(rect=[0, 0.03, 1, 0.95])  # Leave space for main title
plt.show()
```

**The churn rate is higher among customers using **Fiber optic internet**, lacking **tech support**, **online security**, and **backup services**.
Customers **without additional services** (like streaming or multiple lines) tend to churn less.
**Phone service** users show more churn than those without.
Overall, lack of value-added services correlates with higher churn.
Features enhancing support and security appear to improve customer retention.**
